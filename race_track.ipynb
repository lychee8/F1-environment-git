```python
import pandas as pd

df = pd.read_csv("races.csv")

# print(df.head())
# print(df.isnull().sum())
print(df.dtypes)

# print(df.columns)
# df['driverId']

df.drop(['time'], axis = 1, inplace = True)
df.drop(['url'], axis = 1, inplace = True)
df.drop(['fp1_date'], axis = 1, inplace = True)
df.drop(['fp1_time'], axis = 1, inplace = True)
df.drop(['fp2_date'], axis = 1, inplace = True)
df.drop(['fp2_time'], axis = 1, inplace = True)
df.drop(['fp3_date'], axis = 1, inplace = True)
df.drop(['fp3_time'], axis = 1, inplace = True)
df.drop(['quali_date'], axis = 1, inplace = True)
df.drop(['quali_time'], axis = 1, inplace = True)
df.drop(['sprint_date'], axis = 1, inplace = True)
df.drop(['sprint_time'], axis = 1, inplace = True)


# df.columns =df.columns.str.strip()

df['raceId'] = pd.to_numeric(df['raceId'], errors='coerce')

df = df[(df ['raceId'] >= 1121) & (df ['raceId'] <= 1144)] 

# df['time'] = pd.to_numeric(df['time'], errors='coerce')

# df.loc[(df['raceId'] >= 1121) & (df['raceId'] <= 1144), 'raceId'] = 2024

df = df.sort_values(by='raceId', ascending=True)


print(df.head())


df.to_csv("race_track.csv", index=False)

df.describe()
```
